Sandbox for accessing endpoints from nba_api

```python
from tools import show_df

import pandas as pd

from nba_api.stats.endpoints import (
    CommonPlayerInfo,
    AssistTracker,
    AllTimeLeadersGrids,
    CommonAllPlayers,
    CommonTeamRoster,
    CommonTeamYears,
    CumeStatsPlayer,
    CumeStatsPlayerGames,
    CumeStatsTeam,
    CumeStatsTeamGames,
    DraftBoard,
    DraftHistory,
    FranchiseHistory,
    GameRotation,
    LeagueDashPlayerStats,
    LeagueDashTeamStats,
    LeagueLeaders,
    LeagueStandings,
    PlayerAwards,
    PlayerCareerStats,
    PlayerCompare,
    PlayerDashboardByGameSplits,
    PlayerDashboardByGeneralSplits,
    PlayerDashboardByTeamPerformance,
    PlayerDashboardByYearOverYear,
    PlayerGameLog,
    PlayerGameLogs,
    PlayerIndex,
    PlayerProfileV2,
    ScheduleLeagueV2,
    scheduleleaguev2int,
    TeamAndPlayersVsPlayers,
    TeamDashLineups,
    TeamDetails,
    TeamEstimatedMetrics,
    TeamInfoCommon,
    TeamPlayerDashboard,
    TeamPlayerOnOffDetails,
    WinProbabilityPBP
)
```

```python
# Example: Fetch player info for Lebron James
player_id = 2544  # LeBron James' player ID
player_info = CommonPlayerInfo(player_id=player_id)
df = player_info.get_data_frames()[1]
show_df(df)
```

```python
# Example: Fetch top 10 all-time leaders in points per game
advanced_stats = AllTimeLeadersGrids(league_id='00', per_mode_simple='PerGame', season_type='Regular Season', topx=10)
adv_df = advanced_stats.get_data_frames()[1]
show_df(adv_df)
```

```python
# Example: Fetch assist tracker data for Hawks 2024-25 regular season
asts = AssistTracker(per_mode_simple_nullable='Totals', season_nullable='2024-25', league_id_nullable='00', team_id_nullable='1610612737', season_type_all_star_nullable='Regular Season')
asts_df = asts.get_data_frames()[0]
show_df(asts_df)
```

```python
# Example: Fetch roster for the Cleveland Cavaliers for the 2025-26 season
team_id = 1610612739  # Team ID for the Cleveland Cavaliers
season = '2025-26'
roster = CommonTeamRoster(team_id=team_id, season=season)
roster_df = roster.get_data_frames()[0]
show_df(roster_df)
```

```python
# Example: Fetch all current NBA players
all_players = CommonAllPlayers(is_only_current_season=1)
all_players_df = all_players.get_data_frames()[0]
show_df(all_players_df)
```
